### ***as you know in the preprocessing 1 file we already have data of year 2016 and before, so now from meta and credits we will only pull data of year 2017.***

### ***for the data of 2018, 19 and 20 we will be using wikipedia in the future files.***

```python
data['genres'].dtypes
```

```python
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if(i.get('name') == 'Science Fiction'):
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if(gen == []):
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_genresList(x))
```

### what we have tried to do above is to iterate over the object named genres in the data and extract the name property only and keep appending it at the end of an empty list. it the end return NaN if the list is empty of else return a string of this list.

```python
# as you can see now that the new column named genres_list contains a string of genres and not the entire useless data of objects that we didn't want. we will do this for the other two columns also. (cast and crew)
data.head()['genres_list']
```

### this is how the cast of one movie looks like. we will get actor1, actor2, actor3 from this

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if(casts == []):
        return np.NaN
    else:
        return casts[0]
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
data['actor_1_name']
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
data['actor_2_name']
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
data['crew'][1]
```

### this is how the crew column of the data frame looks like. we will iterate over each value and get that crew member whose job type is 'director'

```python
# if there is no director then we are returning nan, else we are returning a string made from all the values in the list that is formed by multiple appends.

def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data['director_name']
```

### now we have all that we wanted. we will now form a new data frame called movie having only those columns that we want.

```python
movie = data.loc[:,['director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'genres_list', 'title']]
```

```python
movie = movie.dropna(how = "any")
```

```python
# just rename our columns to match the column names of the dataset formed earliere because we are going to merge them together

movie = movie.rename(columns={'genres_list': 'genres'})
movie = movie.rename(columns={'title': 'movie_title'})

```

```python
movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
# make a new column combining all the columns into one

movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
```
